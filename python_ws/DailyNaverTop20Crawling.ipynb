```python
def addZero(isLowerTen):
    for index in range(len(isLowerTen)):
        if isLowerTen[index] < 10:
            isLowerTen[index] = '0' + str(isLowerTen[index])
        else:
            isLowerTen[index] = str(isLowerTen[index])
    return isLowerTen
```

# 상장 종목 정보 불러오기

```python
url = 'http://kind.krx.co.kr/corpgeneral/corpList.do'
url += '?currentPageSize=5000'
url += '&fiscalYearEnd=all'
url += '&location=all'
url += '&method=download'
url += '&orderMode=3'
url += '&orderStat=D'
url += '&pageIndex=1'
url += '&searchType=13'
df = pd.read_html(url, header=0)
jongmok_list = list(df[0]['회사명'])
```

## 테이블 : 종목 정보
### row : 종목 이름
### column : 점유 빈도(분), 연관 검색어

```python
df_jongmok = df_ranking[df_ranking.isin(jongmok_list)]
store_list = []

for time in times:
    for jongmok in df_jongmok:
        data = df_jongmok[jongmok][time]
        try:
            math.isnan(data)
        except:
            item_rank = jongmok.split('rank')[1]
            item_time = str(time.time())
            item = [data, item_rank, item_time]
            store_list.append(item)
keys_jongmok = list(set(map(lambda i:i[0], store_list)))
count_jongmok = {i : 0 for i in keys_jongmok}
rel_jongmok = {i : {'관련':[], '비관련':[]} for i in keys_jongmok}
for item in store_list:
    count_jongmok[item[0]] += 1
    
for keyword in keys_jongmok:
    url = 'https://search.naver.com/search.naver'
    params = {
        "sm" : "top_hty",
        "fbm" : "1",
        "ie" : "utf8",
        "query" : keyword
    }
    resp = requests.get(url, params=params)
    soup = bs(resp.text, 'html.parser')
    rel_keywords_area = soup.select_one('.relate_area')
    rel_keywords = []
    if rel_keywords_area == None:
        pass
    else:
        rel_keywords = list(map(lambda i:i.text.strip(), rel_keywords_area.select('li')))
        unrel_keywords = []
        cnt = 0
        for i in range(len(rel_keywords)):
            if rel_keywords[i-cnt] in jongmok_list:
                pass
            else:
                unrel_keywords.append(rel_keywords[i-cnt])
                del rel_keywords[i-cnt]
                cnt += 1
    rel_jongmok[keyword]['관련'] = rel_keywords
    rel_jongmok[keyword]['비관련'] = unrel_keywords
for k in rel_jongmok.keys():
    rel_jongmok[k]['count'] = count_jongmok[k]
```
